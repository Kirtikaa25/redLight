```python
import pandas as pd

# Load the dataset
file_path = "/content/Final_Categorized_Red_Light_Violation.csv"
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
label_encoders = {}
for col in df.columns:
    if df[col].dtype == "object":  # Check if column is categorical
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le  # Store encoder for decoding later
```

```python


from collections import OrderedDict
import pylogit as pl

# Define the target variable (red light violation)
target_column = "Behavioral Factors | Have you ever crossed the street during a red light? (Binary)"

# Define independent variables (features)
X = df.drop(columns=[target_column])
y = df[target_column]

# Convert dataset into long format required by Mixed Logit
long_data = df.melt(id_vars=[target_column], var_name="Factor", value_name="Value")

# Convert specification dictionary into OrderedDict (Required by pylogit)
specification = OrderedDict({col: "random" for col in X.columns})

# Define Mixed Logit model
model = pl.create_choice_model(
    data=long_data,
    alt_id_col="Factor",
    obs_id_col=target_column,
    choice_col="Value",
    specification=specification,  # FIXED: Now using OrderedDict
    model_type="Mixed Logit",
)

# Train the Mixed Logit model
model.fit_mle()


```
